### Add New Columns From Existing Columns

**กำหนดให้**

คะแนนเต็มของแต่ละวิชา
- Math คะแนนเต็ม 50
- Eng คะแนนเต็ม 300
- Sci คะแนนเต็ม 90

เกณฑ์การตัดเกรดคือ
- ในแต่ละวิชา ถ้าได้คะแนน **มากกว่าหรือเท่ากับ** 80% ได้เกรด 4
- ในแต่ละวิชา ถ้าได้คะแนน **มากกว่าหรือเท่ากับ** 70% ได้เกรด 3
- ในแต่ละวิชา ถ้าได้คะแนน **มากกว่าหรือเท่ากับ** 60% ได้เกรด 2
- ในแต่ละวิชา ถ้าได้คะแนน **มากกว่าหรือเท่ากับ** 50% ได้เกรด 1
- ในแต่ละวิชา ถ้าได้คะแนน **ต่ำกว่า** 50% ได้เกรด 0

````python
import numpy as np

data = {
    "Name": ["นาย ก.", "นาย ข.", "นาย ค.", "นาย ง.", "นาย ฉ.", "นาย ช."],
    "Age": [19, 19, 20, 19, 18, 18],
    "Math": [32.4, 22.4, 40.8, 44.0, 25.0, 30.2],
    "Eng": [220, 180, 130, 240, 260, np.nan],
    "Sci": [66, 70, 72, 50, 50, np.nan]
}
df = pd.DataFrame(data)
df
````

````python
df['Math%'] = df['Math'] / 50 * 100
````

````python
df
````

````python
df['Eng%'] = df['Eng'] / 300 * 100
df['Sci%'] = df['Sci'] / 90 * 100
````

เกณฑ์การตัดเกรดคือ
- ในแต่ละวิชา ถ้าได้คะแนน **มากกว่าหรือเท่ากับ** 80% ได้เกรด 4
- ในแต่ละวิชา ถ้าได้คะแนน **มากกว่าหรือเท่ากับ** 70% ได้เกรด 3
- ในแต่ละวิชา ถ้าได้คะแนน **มากกว่าหรือเท่ากับ** 60% ได้เกรด 2
- ในแต่ละวิชา ถ้าได้คะแนน **มากกว่าหรือเท่ากับ** 50% ได้เกรด 1
- ในแต่ละวิชา ถ้าได้คะแนน **ต่ำกว่า** 50% ได้เกรด 0

````python
def calculate_grade(num):
  # Your code here
  if num >= 80:
    return 4
  elif num >= 70:
    return 3
  elif num >= 60:
    return 2
  elif num >= 50:
    return 1
  else:
    return 0
````

เมื่อเราทำการสร้างฟังก์ชันขึ้นมา เราควรทำการเทส (Unit Testing) ฟังก์ชันที่เราสร้างขึ้นมาว่ามันทำงานได้ตามที่เราต้องการหรือไม่ ก่อนนำไปใช้จริง

````python
test_input = [49, 50, 55, 59, 60, 65, 69, 70, 75, 79, 80, 85]
result = map(calculate_grade, test_input)

list(result)
````

นำฟังก์ชัน `calculate_grade` ไปใช้คำนวณเกรดวิชา Math และเก็บผลลัพธ์ที่ได้ใน Column ใหม่ชื่อ `gradeMath` โดยใช้ `apply`

````python
df['gradeMath'] = df['Math%'].apply(calculate_grade)
````

````python
df
````

**ให้คำนวณเกรดวิชา Eng และ Sci**

````python
df['gradeEng'] = df['Eng%'].apply(calculate_grade)
df['gradeSci'] = df['Sci%'].apply(calculate_grade)
````

````python
df
````

**ให้คำนวณเกรดเฉลี่ยของนักเรียนแต่ละคน และเก็บผลลัพธ์ไว้ใน Column ใหม่ชื่อ GPA**

````python
df['GPA'] = (df['gradeMath'] + df['gradeEng'] + df['gradeSci']) / 3
````

## Working with text data
There are two ways to store text data in pandas:

- `object` -dtype NumPy array.
- `StringDtype` extension type.

````python
pd.Series(["a", "b", "c"])
````

````python
pd.Series(["a", "b", "c"], dtype="string") # or dtype=pd.StringDtype()
````

### String methods

Series and Index are equipped with a set of string processing methods that make it easy to operate on each element of the array.

These are accessed via the `str` attribute and generally have names matching the equivalent (scalar) built-in string methods.

The full list of available methods can be found [HERE](https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html#method-summary)

````python
s = pd.Series(
    ["A", "B", "C", np.nan, "CABA", "dog", "cat"], dtype="string"
)
````

````python
s.str.lower()
````

````python
s.str.upper()
````

````python
s.str.len()
````

### Splitting and replacing strings

Methods like `split` return a Series of lists:

````python
s2 = pd.Series(["a_b_c", "c_d_e", np.nan, "f_g_h"], dtype="string")
s2.str.split("_")
````

Elements in the split lists can be accessed using `get` or `[]` notation:

````python
print(s2.str.split("_").str.get(1))
print(s2.str.split("_").str[0])
````

It is easy to expand this to return a DataFrame using `expand` option.

````python
s2.str.split("_", expand=True)
````

````python
s3 = pd.Series(
    ["A", "B", "C", "Aaba", "Baca", "", np.nan, "CABA", "dog", "cat"],
    dtype="string",
)
````

````python
s3.str.replace('A', 'XX', regex=False)
````

````python
# Optionally replace can use Regex
s3.str.replace("^.a|dog", "XX-XX ", case=False, regex=True)
````

### Concatenation

````python
s = pd.Series(["a", "b", "c", "d"], dtype="string")
print(s)
s.str.cat(sep=",")
````

By default, missing values are ignored. Using na_rep, they can be given a representation:

````python
t = pd.Series(["a", "b", np.nan, "d"], dtype="string")
# t.str.cat(sep=",")
t.str.cat(sep=",", na_rep="-")
````

Concatenating a Series and something list-like into a Series.

The first argument to cat() can be a list-like object, provided that it matches the length of the calling Series.

````python
s
````

````python
s.str.cat(others=["A", "B", "C", "D"])
````

**EX:** Concatenating address with name column using `', '` for separator

````python
df['Name'] = df['Name'] + ', ' + df['Address']
df
````

````python
# YOUR CODE HERE
df['Name'] = df['Name'].str.cat(sep=', ', others=df['Address'])
df
````
