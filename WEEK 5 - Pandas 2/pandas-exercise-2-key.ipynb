**Which is the most age of dead passenger? (frequency)**

```python
# YOUR CODE HERE
df[df['Survived'] == 0]['Age'].value_counts()
```

**Average age of dead female passengers and dead male passengers**

```python
# YOUR CODE HERE
female_age = df[(df['Sex'] == 'female') & (df['Survived'] == 0)]['Age'].mean()
male_age = df[(df['Sex'] == 'male') & (df['Survived'] == 0)]['Age'].mean()

print(f'Female: {female_age} years')
print(f'Male: {male_age} years')
```

**Which columns has null value ? If it has null value how many null rows in each columns**

```python
# YOUR CODE HERE
df.isnull().sum()
```

**In Age column replace null value with average value and then check null again**

```python
# YOUR CODE HERE
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

**Average Ticket price of each Pclass ?**

```python
# YOUR CODE HERE
df.groupby('Pclass')['Age'].mean()
```

**Average Ticket price of each Pclass that passenger died ?**

```python
# YOUR CODE HERE
avg_fare_dead = df[df['Survived'] == 0].groupby('Pclass')['Fare'].mean()
avg_fare_dead
```

**Average Ticket price of each Pclass that passenger survived ?**

```python
# YOUR CODE HERE
avg_fare_survive = df[df['Survived'] == 1].groupby('Pclass')['Fare'].mean()
avg_fare_survive
```

**Average Ticket price of each Pclass, calculated by all pasengers, dead passengers, and survived passengers ?**

```python
# YOUR CODE HERE
avg_fare_all = df.groupby('Pclass')['Fare'].mean()
avg_fare_all
```

```python
new_df = pd.concat([avg_fare_dead, avg_fare_survive, avg_fare_all], axis=1)
new_df.columns = ["Avg Dead", "Avg Survive", "Avg All"]
new_df
```

**Which port has the best rate survived?**

C = Cherbourg, Q = Queenstown, S = Southampton

```python
# YOUR CODE HERE
df.groupby(['Survived', 'Embarked'])['Embarked'].count()
```

```python
s1 = df.groupby(['Embarked'])['Embarked'].count()

def calSurvivedPercent(x):
  return len(x[x['Survived'] == 1]) / len(x) * 100

s2 = df.groupby(['Embarked']).apply(calSurvivedPercent)

new_df = pd.concat([s1, s2], axis=1)
new_df.columns = ['Count', 'Percent Survived']
new_df
```
